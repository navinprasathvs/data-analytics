```python
# Visualization 2: Scatter Plot - Relationship between kilometers driven and price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='kms_driven', y='price', hue='brand', data=cleaned_data, palette='tab10')
plt.title('Kilometers Driven vs Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.legend(title='Brand', bbox_to_anchor=(1, 1))
plt.show()

```

```python
# Visualization 3: Histogram - Distribution of bike ages
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['age'], kde=True, bins=10, color='skyblue')
plt.title('Distribution of Bike Ages')
plt.xlabel('Age (Years)')
plt.ylabel('Frequency')
plt.show()

```

```python
# Visualization 4: Box Plot - Distribution of prices by ownership status
plt.figure(figsize=(10, 6))
sns.boxplot(x='owner', y='price', data=cleaned_data, palette='pastel')
plt.title('Price Distribution by Ownership Status')
plt.xlabel('Ownership Status')
plt.ylabel('Price')
plt.show()

```

```python
# Visualization 5: Pie Chart - Proportion of bikes by city
city_counts = cleaned_data['city'].value_counts()
plt.figure(figsize=(8, 8))
city_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('Set3', len(city_counts)))
plt.title('Proportion of Bikes by City')
plt.ylabel('')  # Hide the y-label for better aesthetics
plt.show()
```

```python
# Visualization 6: Heatmap - Correlation matrix between numeric columns
plt.figure(figsize=(10, 6))
correlation_matrix = cleaned_data[['price', 'kms_driven', 'age', 'power']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
# Visualization 7: Count Plot - Number of bikes by ownership status
plt.figure(figsize=(10, 6))
sns.countplot(x='owner', data=cleaned_data, palette='Blues')
plt.title('Number of Bikes by Ownership Status')
plt.xlabel('Ownership Status')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 8: Violin Plot - Price distribution across brands
plt.figure(figsize=(12, 6))
sns.violinplot(x='brand', y='price', data=cleaned_data, palette='muted', scale='width')
plt.title('Price Distribution Across Brands')
plt.xticks(rotation=45)
plt.xlabel('Brand')
plt.ylabel('Price')
plt.show()

```

```python
# Visualization 9: Line Plot - Average price trend with bike age
age_price_trend = cleaned_data.groupby('age')['price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='price', data=age_price_trend, marker='o', color='orange')
plt.title('Average Price Trend by Bike Age')
plt.xlabel('Age (Years)')
plt.ylabel('Average Price')
plt.show()
```

```python
# Visualization 10: Stacked Bar Chart - Bike counts by brand and ownership status
brand_owner_counts = cleaned_data.groupby(['brand', 'owner']).size().unstack(fill_value=0)
brand_owner_counts.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title('Bike Counts by Brand and Ownership Status')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.legend(title='Ownership Status')
plt.xticks(rotation=45)
plt.show()
```
